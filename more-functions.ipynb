### Track the Early and Late Intervals of Species and Genus along with Mean Age

```python
infile = "/home/eeb177-student/Desktop/eeb-177/final-project/chiropteradataset.csv"
with open(infile,'r') as bat_records:
        trial = bat_records.readlines()[17:]
```

```python
## make a dictionary
from collections import defaultdict
age_range = defaultdict(list)

output = open("bat-age-ranges.csv", "w")

for line in trial:
        species = (line.split('","')[5]) #line[5] would only give 5th character
        min_age = line.split('","')[15] #can also strip quotes with minage.linesplit(",".strip('""')[5])
        max_age = line.split('","')[14]
        mean_age = (float(min_age) + float(max_age))/2
        #print(meanage)
        
        age_range[species].append(mean_age)
        print(mean_age)
```

```python
output = open("chiropteradata.csv", "w")

for line in trial:
        species = line.split('","')[5] 
        early_int = line.split('","')[12]
        late_int =  line.split('","')[13]
        genus = species.split(" ")[0]

        output = "{},{},{},{}\n".format(genus,species,early_int,late_int)
        print(output)
        
## since I want to plot species diversity through time based on initial appearance,
#I will use this to extract the species with time interval
#I can also use this to look at genus and early time interval 
#may not need late time interval since bats are not going extinct????
```
