```python
infile = "/home/eeb177-student/Desktop/eeb-177/final-project/chiroptera/newchiropteraspecies.csv"
with open(infile,'r') as species_records:
        trial = species_records.readlines()[17:]
```

```python
output = open("chiropteraspeciesonly.csv", "w")

for line in trial:
        name = line.split('","')[5]
        genus = name.split(" ")[0]
        species = name.split(" ")[1]

        outline = "{},{}\n".format(genus,species)
        output.write(outline)
        print(outline)
        
output.close()
```

```python
## make a dictionary
from collections import defaultdict
age_range = defaultdict(list)

output = open("chiroptera-age-ranges.csv", "w")

for line in trial:
        species = (line.split('","')[5]) #line[5] would only give 5th character
        min_age = line.split('","')[11] #can also strip quotes with minage.linesplit(",".strip('""')[5])
        max_age = line.split('","')[10]
        mean_age = (float(min_age) + float(max_age))/2
        #print(meanage)
        
        age_range[species].append(mean_age)
        print(mean_age)
```
